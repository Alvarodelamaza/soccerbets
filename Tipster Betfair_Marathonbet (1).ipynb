```python
import sys
import csv
from selenium import webdriver
import time
import numpy as np
import pandas as pd
from selenium.webdriver.common.keys import Keys
from datetime import date
from datetime import datetime
from selenium.webdriver.support.select import Select

#Betfair
url = "https://www.betfair.es/exchange/plus/en/football-betting-1"
## Add in the next line your path to the Chromedriver (the one that is now its mine)
path = '/Users/alvarodelamaza/Downloads/chromedriver 6'

myOptions=webdriver.ChromeOptions()
myOptions.add_argument('--disable-infobars')
myOptions.add_argument('--disable-extensions')
myOptions.add_argument('--no-sandbox')
myOptions.add_argument('--disable-popup-blocking')
myOptions.add_argument('--disable-plugins-discovery')
#myOptions.add_argument('--headless')
driver = webdriver.Chrome(path,options=myOptions)

driver.get(url)

time.sleep(6)
cookies=driver.find_element_by_xpath(".//button[@id='onetrust-accept-btn-handler']")
driver.execute_script("arguments[0].click();", cookies)
time.sleep(3)

k=0
total=[]
horah=[]
horahm=[]
num_page=1000
while k< num_page:
    k=k+1
    horaa=[] 
    casa=[]
    visit=[]
    hora=[]
    equipos=[]
    oneback=[]
    xback=[]
    twoback=[]
    onelay=[]
    xlay=[]
    twolay=[]
    match=[]
    fecha=[]
    identity=[]
    casaa=[]
    visitt=[]
    equipos=(driver.find_elements_by_xpath(".//ul[@class='runners']"))
    
    horarios=driver.find_elements_by_xpath(".//a[@class='market-rules-icon']")
    for t in range(len(horarios)):
        horarios[t].click()
        time.sleep(0.25)
        horaa=driver.find_elements_by_xpath(".//span[@class='market-rules-event-start-time']")
        hora.append(horaa[0].text)
        (driver.find_elements_by_xpath(".//button[@class='ngdialog-close']")[0]).click()
    
    for i in range(len(equipos)):
        
        match=equipos[i].text
        match=match.replace('\n',',')
        sep=match.split(',')
        casa.append(sep[0])
        visit.append(sep[1])
        casaa=casa[i].replace(' ','')
        visitt=visit[i].replace(' ','')
        hora[i]=hora[i].replace(',','-')
        hora[i]=hora[i].replace(' ','')
        hora[i]=hora[i].replace(' ','')
        hora[i]=hora[i].replace('November','Nov')
        hora[i]=hora[i].replace('December','Dec')
        hora[i]=hora[i].replace('January','Jan')
        hora[i]=hora[i].replace('February','Feb')
        hora[i]=hora[i].replace('March','Mar')
        hora[i]=hora[i].replace('April','Apr')
        hora[i]=hora[i].replace('May','May')
        hora[i]=hora[i].replace('June','Jun')
        hora[i]=hora[i].replace('July','Jul')
        hora[i]=hora[i].replace('August','Aug')
        hora[i]=hora[i].replace('September','Sep')
        hora[i]=hora[i].replace('October','Oct')
        identit=casaa+visitt+hora[i]
        identity.append(identit)
        
    
    numbers=(driver.find_elements_by_xpath(".//span[@class='bet-button-price']"))
    for m in range(0,len(numbers),6):
        oneback.append(numbers[m].text)
        onelay.append(numbers[m+1].text)
        xback.append(numbers[m+2].text)
        xlay.append(numbers[m+3].text)
        twoback.append(numbers[m+4].text)
        twolay.append(numbers[m+5].text)
        
    for k in range(len(equipos)):
        item=[]
        horah.append(hora[k][0:-5])
        item.append(identity[k])
        item.append(hora[k])
        item.append(casa[k])
        item.append(visit[k])
        item.append(oneback[k])
        item.append(onelay[k])
        item.append(xback[k])
        item.append(xlay[k])
        item.append(twoback[k])
        item.append(twolay[k])
        total.append(item)
    
    time.sleep(1)
    try:
        nextt=(driver.find_elements_by_xpath(".//a[@class='coupon-page-navigation__link coupon-page-navigation__link--next']"))
        nextt[0].click()
        time.sleep(1)
    except:
        break
        
tipster=pd.DataFrame(total,columns=['ID','StartDate','HomeTeam','AwayTeam','BF1B','BF1L','BFXB','BFXL','BF2B','BF2L'])  


#Marathonbet

driver.get('https://www.marathonbet.es/es/popular/Football+-+11?interval=ALL_TIME')

cookies=driver.find_element_by_xpath(".//button[@class='v-btn v-btn--is-elevated v-btn--has-bg v-size--small v-btn--primary']")
driver.execute_script("arguments[0].click();", cookies)
time.sleep(2)
lenguaje=driver.find_element_by_xpath(".//span[@class='font-icon marathon_icons-COUNTRY_ES language__flag']")
driver.execute_script("arguments[0].click();", lenguaje)
time.sleep(2)
lenguajee=driver.find_element_by_xpath(".//span[@class='language__flag font-icon marathon_icons-COUNTRY_GB']")
driver.execute_script("arguments[0].click();", lenguajee)
time.sleep(3)
teamss=[]
webgo=[]
target = driver.find_element_by_xpath(".//span[@class='footer-menu__item']")
driver.execute_script('arguments[0].scrollIntoView(true);', target)

time.sleep(3)

teams=driver.find_elements_by_xpath(".//a[@class='member-link']") 
for k in range(len(teams)):
    webgo.append(teams[k].get_attribute('href'))
    
onem=[]
xm=[]
twom=[]
when=[]
totalm=[]
teamssa=[]
teamssc=[]
cuando=[]
numberm1=[]
numberm=[]
identitym=[]

for i in range(0,len(teams),2):
    teamssc.append(teams[i].text)
    teamssa.append(teams[i+1].text)
    
for i in range(len(teamssc)):
    time.sleep(2)
    totnum=[]
    driver.get(webgo[i+i])
    numberm1.append(driver.find_elements_by_xpath(".//td[@class='price height-column-with-price    first-in-main-row  coupone-width-3']"))
    numberm.append(driver.find_elements_by_xpath(".//td[@class='price height-column-with-price    coupone-width-3']"))
    try:
        totnum.append(numberm1)
    except:
        pass
    try:
        totnum.append(numberm[i][0])
        totnum.append(numberm[i][1])
    except:
        pass
    
    if len(totnum)<3:
        xm.append('-')
        twom.append('-')
        onem.append('-')
    else:
        xm.append(numberm[i][0].text)
        twom.append(numberm[i][1].text)
        onem.append(numberm1[i][0].text)
    try:
        schedule=driver.find_elements_by_xpath('.//td[@class=" date date-short"]')
        cuando.append(schedule[0].text)
    except:
        schedule1=driver.find_elements_by_xpath('.//td[@class=" no-zero-margin-td date date-with-month"]')
        cuando.append(schedule1[0].text)
        
    
    time.sleep(0.1)
today = date.today()
for k in range(len(cuando)):
    if len(cuando[k])==5:
        cuando[k]= today.strftime("%d%b ")+cuando[k]
        cuando[k]=cuando[k].replace(' ','%s-' %today.strftime("%Y"))
    else:
        cuando[k]=cuando[k].replace(' ','',1)
        cuando[k]=cuando[k].replace(' ','%s-' %today.strftime("%Y"))
    identitym.append(teamssc[k]+teamssa[k]+cuando[k])
    identitym[k]=identitym[k].replace(' ','')
        
for j in range(len(teamssc)):
    itemm=[]
    horahm.append(cuando[j][0:-5])
    itemm.append(identitym[j])
    itemm.append(cuando[j])
    itemm.append(teamssc[j])
    itemm.append(teamssa[j])
    itemm.append(onem[j])
    itemm.append(xm[j])
    itemm.append(twom[j])
    totalm.append(itemm)


driver.quit()  
   
tipsterm=pd.DataFrame(totalm,columns=['ID','StartDate','HomeTeam','AwayTeam','MB1','MBX','MB2'])  
totalcon=[]

for m in range(len(totalm)):
    for b in range(len(total)):
        equal=[]
        
        if (horahm[m]==horah[b] and totalm[m][3]==total[b][3]) or (totalm[m][3]==total[b][3] and totalm[m][2]==total[b][2]) or (totalm[m][2]==total[b][2] and horahm[m]==horah[b]) :
            equal.append(total[b][0])
            equal.append(total[b][1])
            equal.append(total[b][2])
            equal.append(total[b][3])
            equal.append(total[b][4])
            equal.append(total[b][5])
            equal.append(total[b][6])
            equal.append(total[b][7])
            equal.append(total[b][8])
            equal.append(total[b][9])
            equal.append(totalm[m][4])
            equal.append(totalm[m][5])
            equal.append(totalm[m][6])
            try:
                equal.append('%.3f' % (float(totalm[m][4])-float(total[b][5])))
            except:
                equal.append(0)
            try:
                equal.append('%.3f' % (float(totalm[m][5])-float(total[b][7])))
            except:
                equal.append(0)
            try:
                equal.append('%.3f' % (float(totalm[m][6])-float(total[b][9])))
            except:
                equal.append(0)
            try:
                maxdiff=max((float(totalm[m][4])-float(total[b][5])),(float(totalm[m][5])-float(total[b][7])),(float(totalm[m][6])-float(total[b][9])))
                equal.append('%.3f' % float(maxdiff))
            except:
                maxdiff='n/a'
                equal.append(maxdiff)
           
            totalcon.append(equal)
tipstermm=pd.DataFrame(totalcon,columns=['ID','StartDate','HomeTeam','AwayTeam','BF1B','BF1L','BFXB','BFXL','BF2B','BF2L','MB1','MBX','MB2','1DIFF','XDIFF','2DIFF','MAXDIFF'])  


writerr = pd.ExcelWriter('Scraping Odds %s.xlsx' % today.strftime("%d%b%Y"))
tipstermm.to_excel(writerr)
writerr.save()
print('DataFrame is written successfully to Excel File.')

```
